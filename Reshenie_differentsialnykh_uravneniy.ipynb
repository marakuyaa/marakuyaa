### Метод Рунге Кутты
#### Четвёртый порядок сходимости

```python
import math as mt
import numpy as np
import matplotlib.pyplot as plt
```

```python
def runge_kutta(f, u0, t0, tN, h):
    N = int((tN - t0) / h)
    u = np.zeros(N)
    u[0] = u0
    for n in range (0, N-1):
        tn = t0 + n * h
        k1 = h * f(tn , u[n])
        k2 = h * f(tn + h/2, u[n] + k1/2)
        k3 = h * f(tn + h/2, u[n] + k2/2)
        k4 = h * f(tn + h, u[n] + k3)
        u[n+1] = u[n] + 1/6 * (k1 + 2*k2+ 2*k3 +k4)
    return u
```
